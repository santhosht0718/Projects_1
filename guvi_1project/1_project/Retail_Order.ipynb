```python
import pandas as pd
# load data and reading
df=pd.read_csv("D:\guvi_1project\orders (1).csv")   
df
```

```python
# This is useful for quickly checking the structure of your dataset.
df.info()
```

```python
# used to check how many missing (Null) values each column has.
df.isnull().sum()
```

```python
# Remove duplicate rows from a DataFrame.
df.drop_duplicates(inplace=True)
```

```python
# renaming your DataFrame columns for consistency!
df.rename(columns={'Order Id':'order_id','Order Date':'Order_Date','Ship Mode':'Ship_Mode','Postal Code':'Postal_Code','Sub Category':'Sub_Category','Product Id':'Product_Id',
'cost price':'cost_price','List Price':'List_Price','Discount Percent':'Discount_Percent'},inplace=True)
df.head(5)
```

```python
# checking the percentage of missing values in each column.

df1 = (df.isnull().sum()/len(df))*100
df1[df1>0]
```

```python
# fill missing values in categorical columns using the mode

df["Ship_Mode"]=df["Ship_Mode"].fillna(df["Ship_Mode"].mode()[0])
```

```python
df1 = (df.isnull().sum()/len(df))*100
df1[df1>0]
```

```python
# new columns add discount , sale price , profit

df[' discount']=df['List_Price']*df['Discount_Percent']*.01
df['sale_price']=df['List_Price']-df[' discount']
df['profit']=df['sale_price']-df['cost_price']
df.head(5)
```

```python
# converting your Order_Date column to a datetime data type

df['Order_Date']=pd.to_datetime(df['Order_Date'])
```
